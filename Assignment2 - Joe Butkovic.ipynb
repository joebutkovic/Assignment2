```python

balance = float(input('Enter the principal loan amount: '))
loanAmt = balance
interestRate = float(input('Enter the interest rate: '))
monthlyPayment = float(input('Enter the monthly payment: '))
month = 1
interest = (balance - monthlyPayment)*(interestRate/12)
totalInterest = 0.0

if balance*(interestRate/12)>monthlyPayment:
    print("YOU WILL NEVER PAY OFF YOUR LOAN.")
else:
    print('Month', end='\t\t')
    print('Payment', end='\t\t')
    print('Interest', end='\t')
    print('Balance')
    while balance > 0:
        print(month, end='\t\t')
        if balance < monthlyPayment:
            print("%.2f" % round(balance,2), end='\t\t')
        else:
            print("%.2f" % round(monthlyPayment,2), end='\t\t')
        interest = balance*(interestRate/12)
        balance = balance - (monthlyPayment - interest)
        print("%.2f" % round(interest,2), end='\t\t')
        if balance < 0:
            print("0.00")
        else:
            print("%.2f" % round(balance,2))    
        month = month + 1
        totalInterest = totalInterest + interest
    totalAmtPaid = totalInterest + loanAmt
    print('Total Interest Accrued: '+"%.2f" % round(totalInterest,2))
    print('Total Amount Paid: '+"%.2f" % round(totalAmtPaid,2))
    totalTimeYRS = int((month-1) / 12)
    totalTimeYRSMONTHS = (month-1) % 12
    print('Total time to pay off this loan:',end=' ')
    print(totalTimeYRS,end=' ')
    print('years and',end=' ')
    print(totalTimeYRSMONTHS,end=' ')
    print('months',end=' ') 
```
